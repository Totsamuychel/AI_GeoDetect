Цей ноутбук візуалізує географічний розподіл завантажених зображень України та сусідніх країн.

```python
import pandas as pd
import plotly.express as px
import os

# Завантаження маніфесту (використовуємо шлях відносно папки notebooks)
manifest_path = "../dataset/raw/osv5m/manifest.csv"
if not os.path.exists(manifest_path):
    # Fallback якщо запускаємо з кореня
    manifest_path = "dataset/raw/osv5m/manifest.csv"

df = pd.read_csv(manifest_path)
print(f"Всього зображень: {len(df)}")
df.head()
```

```python
# Статистика за країнами
country_counts = df['country'].value_counts().reset_index()
country_counts.columns = ['Країна', 'Кількість']
fig_pie = px.pie(country_counts, values='Кількість', names='Країна', title='Розподіл за країнами')
fig_pie.show()
```

```python
# Візуалізація на інтерактивній карті
fig = px.scatter_mapbox(df, 
                        lat="lat", 
                        lon="lon", 
                        color="country", 
                        hover_name="image_id",
                        zoom=4, 
                        height=700, 
                        mapbox_style="open-street-map")

fig.update_layout(margin={"r":0,"t":40,"l":0,"b":0}, title='Географічне покриття датасету')
fig.show()
```
